```python
def calculate_bmi(weight_kg,height_m):
  bmi = weight_kg / (height_m ** 2)
  if bmi < 18.5:
        print("Underweight")
  elif 18.5 <= bmi < 25:
        print("Normal")
  else:
        print("Overweight")
  return bmi

print(f"BMI for 70kg, 1.75m: {calculate_bmi(70, 1.75):.2f}")
print(f"BMI for 50kg, 1.75m: {calculate_bmi(50, 1.75):.2f}")
print(f"BMI for 80kg, 1.70m: {calculate_bmi(80, 1.70):.2f}")
```
