```python
from typing import List, Generator, Any, Dict
from functools import wraps

# 1. Decorator
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        print("Starting training...")
        result = func(*args, **kwargs)
        print("Training complete.")
        return result
    return wrapper


# 2. train_model function
@timer
def train_model(model_name: str, *datasets: str, **hyperparams: Any) -> None:
    print(f"Model: {model_name}")
    print(f"Datasets: {datasets}")
    print(f"Hyperparameters: {hyperparams}")


# 3. filter_scores
def filter_scores(scores: List[float]) -> List[float]:
    filtered = filter(lambda x: x > 80, scores)
    rounded = map(lambda x: round(x, 2), filtered)
    return list(rounded)


# 4. Generator
def score_generator(scores: List[float]) -> Generator[float, None, None]:
    for score in scores:
        if score > 70:
            yield score


# 5. save_results
def save_results(model_name: str, scores: List[float]) -> None:
    try:
        with open("results.txt", "w") as f:
            f.write(f"Model: {model_name}\n")
            f.write(f"Scores: {scores}\n")
        print("Results saved to results.txt")
    except FileNotFoundError:
        print("Error: File path not found")


# ---------------- TEST ----------------

# Training
train_model(
    "BERT",
    "dataset1", "dataset2", "dataset3",
    lr=0.001, epochs=10, batch_size=32
)

# Scores
scores = [65.2, 85.5, 92.3, 70.0, 88.7]

filtered = filter_scores(scores)
print("Filtered scores:", filtered)

# Generator usage
print("Scores above 70:", end=" ")
for s in score_generator(filtered):
    print(s, end=" ")
print()

# Save results
save_results("BERT", filtered)
```
