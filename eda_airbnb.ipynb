```python
df = df.dropna(axis=1)
df.isnull().sum()
```

```python
print("\nMISSING VALUE PERCENTAGE")

missing_percent = (df.isnull().sum() / len(df)) * 100

print(missing_percent)
```

```python
print("\nDUPLICATE ROWS")

duplicates = df.duplicated().sum()

print("Total Duplicates:", duplicates)
```

```python
print("\nSHAPE BEFORE REMOVING DUPLICATES")
print(df.shape)

df = df.drop_duplicates()

print("\nSHAPE AFTER REMOVING DUPLICATES")
print(df.shape)

print("\nDuplicates Removed Successfully")
```

```python
print("PRICE DISTRIBUTION")
print("mean",df['price'].mean())
print("median",df['price'].median())
print("min",df['price'].min())
print("max",df['price'].max())
print("std",df['price'].std())
print()
```

```python
print("histogram of price")
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=30)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
print("boxplot of price")
plt.figure(figsize=(8,5))
sns.boxplot(x=df["price"])
plt.title("Price Boxplot")
plt.show()
```

```python
Q1 = df["price"].quantile(0.25)
Q3 = df["price"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[
    (df["price"] < lower_bound)
    |
    (df["price"] > upper_bound)
]
print("OUTLIERS")
print("Total Outliers:", len(outliers))
print()
```

```python
print("SKEWNESS")
skew_value = df["price"].skew()
print("Skewness:", skew_value)

if skew_value > 0:
    print("Price is Right Skewed")
elif skew_value < 0:
    print("Price is Left Skewed")
else:
    print("Price is Symmetric")
```
